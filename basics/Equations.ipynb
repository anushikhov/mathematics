```python
from IPython.display import display, Math

display('4 + 3 = 7')
display('4 + 3 = ' + str(4+3) )
```

```python
display(Math('4+3=7'))
```

```python
x = 4
y = 5

display(Math(str(x) + '+' + str(y) + '=' + str(x+y)))
```

```python
display(Math('%g+%g = %g' %(x, y, x+y)))
```

```python
display(Math('4/5 = .8'))
display(Math('\\frac{4}{5} = .8'))
```

```python
x = 3.4
y = 17

display(Math('%g \\times %g = %g ' %(x, y, x*y)))
```

```python
x = 7
y = -2
z = 5

display(Math(str(3) + 'x(' + str(4) + ' + ' + 'y) = ' + str(42)))
```

```python
display(Math('3\\times%g(4%g) = 42' %(x, y)))
```

```python
display(Math('-y-\\frac{x+3}{z} = 0'))
```

```python
display(Math('-%g-\\frac{%g+3}{%g} = 0' %(x, y, z)))
```
