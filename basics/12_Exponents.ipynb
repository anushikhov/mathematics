```python
from IPython.display import display, Math

# Law of exponents
display(Math('3^2\\times3^4 = 3^{2+4}'))
```

```python
x = 5; y = 5.1

ans = x**(3/4) * 4**y

display(Math('x^{3/4}\\times4^y = %g' %ans))
```

```python
ans = 3**3/x**y

display(Math('\\frac{3^3}{x^y} = %g' %ans))
```

```python
ans = 10**(x-4)

display(Math('10^{x-4}=%g'%ans))
```
